```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("diabetes.csv")
dataset.head()
```

```python
dataset.shape
```

```python
dataset['Outcome'].value_counts()
```

```python
dataset.duplicated().sum()
```

```python
majority_class = dataset[dataset['Outcome'] == 0]  # Majority class (0)
minority_class = dataset[dataset['Outcome'] == 1]  # Minority class (1)
```

```python
majority_class_sampled = majority_class.sample(n=268, random_state=42)
```

```python
dataset = pd.concat([majority_class_sampled, minority_class], ignore_index=True)
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
x= dataset.drop(columns='Outcome' , axis=1)
y = dataset['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train.shape , y_train.shape , X_test.shape , y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
train_scaled = mm.fit_transform(X_train)
test_scaled = mm.transform(X_test)
```

```python
train_scaled
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Input,Dense , Flatten , Conv2D , Dropout
from keras.callbacks import EarlyStopping , ModelCheckpoint , ReduceLROnPlateau
```

```python
model= keras.Sequential([
    keras.layers.Flatten(input_shape=(8,)),
    keras.layers.Dense(128 , activation='relu'),
    Dropout(0.2),
    keras.layers.Dense(64 , activation='relu'),
    Dropout(0.2),
    keras.layers.Dense(32 , activation='relu'),
    Dropout(0.2),
    keras.layers.Dense(2 , activation='softmax')
])

model.summary()
```

```python
es = EarlyStopping(monitor="val_loss", verbose=1, mode="min", patience=1)
cp = ModelCheckpoint("model.keras", monitor="val_loss", save_best_only=True, save_weights_only=False, mode="auto", save_freq="epoch")
lrr = ReduceLROnPlateau(monitor="val_loss", patience=3, verbose=1, factor=0.5, min_lr=0.0001)


model.compile(
    optimizer='adam',
    loss = 'sparse_categorical_crossentropy',
    metrics=['accuracy'])
    
```

```python
model.fit(train_scaled , y_train , epochs=100  , callbacks=[es, cp , lrr])
```

```python
model.evaluate(test_scaled , y_test)
```

```python
import numpy as np
input_text = '1,85,66,29,0,26.6,0.351,31	'
input_arr = np.asarray([float(i) for i in input_text.split(',')]).reshape(1, -1)
input_reshape = mm.transform(input_arr)
result = model.predict(input_reshape)
if result.shape[1] > 1:
    predicted_class = np.argmax(result[0]) 
else:
    predicted_class = int(result[0])  
if predicted_class == 0:
    print("Prediction: Class 0")
else:
    print("Prediction: Class 1")
```
